```python
from autoscraper import AutoScraper
```

```python
myupchar_url = 'https://www.flipkart.com/mobiles/pr?sid=tyy,4io&p[]=facets.brand%255B%255D%3DPOCO&otracker=clp_metro_expandable_2_3.metroExpandable.METRO_EXPANDABLE_POCO_mobile-phones-store_O1WYX08RHODP_wp3&fm=neo%2Fmerchandising&iid=M_45e94375-a56b-4b5a-8d37-34b1bd290069_3.O1WYX08RHODP&ppt=hp&ppn=homepage&ssid=y28b2o7hps0000001627980809306'
```

```python
wanted_list = ['POCO M3 (Power Black, 64 GB)','₹11,499']
```

```python
scraper = AutoScraper()
result = scraper.build(myupchar_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(myupchar_url,grouped=True)
```
